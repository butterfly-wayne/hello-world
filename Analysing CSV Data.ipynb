# Analysing CSV data
This notebook is for outlining how to, and practicing analysing csv data in python.

```python
#Pandas seems to be universally acknowledged as the best way
import pandas as pd
```

```python
#Read in a csv using pd.read_csv
covid=pd.read_csv("C:/Users/user/github/COVID-19/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Deaths.csv")
```

```python
#Look at just the countries
countries=covid['Country/Region']
countries
```

```python
#Set the index and then select by row, selecting just UK
covid.set_index('Country/Region',inplace=True)
covid.loc['United Kingdom']
```

```python
#Select just a row and a column, e.g USA as of 16/03
USA=covid.loc['US',['Province/State','3/16/20']]
USA
```

```python
#For the USA what's the total and average deaths per state - as of 16/3, there are 0.3 deaths per province, there are 247 provinces
print('Total deaths in USA are %u'  %USA["3/16/20"].sum() )
print("Average deaths across the 247 provinces are %u" %USA["3/16/20"].mean())
```
